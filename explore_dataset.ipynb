```python
import matplotlib.pyplot as plt
```

```python
# Counting occurrences of each label
label_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
label_counts = dataset[label_columns].sum()

plt.figure(figsize=(5,3))
plt.bar(label_counts.index, label_counts.values, color='skyblue')
plt.xlabel('Etichette')
plt.ylabel('Conteggio')
plt.title('Conteggio per ciascuna etichetta')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
explosion = [0,0, 0, 0.2,0,0]
plt.figure(figsize=(5,3))
plt.pie(label_counts.values, labels=label_columns, autopct="%.2f%%")
plt.title('Distribuzione delle Etichette')
plt.show()
```

```python
import seaborn as sns
```

```python
# Supponendo che df contenga colonne binarie per ciascuna etichetta
co_occurrence_matrix = np.dot(dataset[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].T, 
                              dataset[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']])

# Creare una heatmap della matrice di co-occorrenza
sns.heatmap(co_occurrence_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'],
            yticklabels=['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'])
plt.title('Matrice di Co-Occorrenza delle Etichette')
plt.show()
```

```python
from sklearn.metrics import jaccard_score
```

```python
# Calcolo dell'indice di Jaccard per ogni coppia di etichette
labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
jaccard_matrix = np.zeros((len(labels), len(labels)))

for i, label1 in enumerate(labels):
    for j, label2 in enumerate(labels):
        if i != j:
            jaccard_matrix[i, j] = jaccard_score(dataset[label1], dataset[label2])

# Visualizzazione dell'indice di Jaccard con una heatmap
sns.heatmap(jaccard_matrix, annot=True, cmap='Purples', xticklabels=labels, yticklabels=labels)
plt.title('Indice di Jaccard tra le Etichette')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Generare le regole di associazione
frequent_itemsets = apriori(dataset[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']], min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
